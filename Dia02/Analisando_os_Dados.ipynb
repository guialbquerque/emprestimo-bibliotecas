#### Excluindo a colina 'registro_sistema', pois não é relevante para a análise

#### Criando nova coluna a partir da coluna 'localizacao', onde a numeracao obedece a seguinte a classificacao:
#### 1. de 0 a 99: 'Generalidades. Ciência e conhecimento'
#### 2. de 100 a 199: 'Filosofia. Psicologia.'
#### 3. de 200 a 299: 'Religião.'
#### 4. de 300 a 399: 'Ciências sociais.'
#### 5. de 400 a 499: 'Classe vaga. Provisoriamente não ocupada.'
#### 6. de 500 a 599: 'Matemática e ciências naturais'.
#### 7. de 600 a 699: 'Ciências aplicadas'
#### 8. de 700 a 799: 'Belas artes'
#### 9. de 800 a 899: 'Linguagem. Língua. Linguística'
#### 10. de 900 a 999: 'Geografia. Biografia. História'


```python
df['CDU'] = df.localizacao.map(lambda x: 'Generalidades. Ciência e conhecimento' if x < 100 else (
    'Filosofia e Psicologia.' if x < 200 else ('Religião' if x < 300 else ('Ciências sociais' if x < 400 else ('Classe vaga. Provisoriamente não ocupada' if 
                            x < 500 else ('Matemática e ciências naturais' if x < 600 else ('Ciências aplicadas' if x < 700 else ( 
                           'Belas artes' if x < 800 else ('Linguagem. Língua. Linguística' if x < 900 else ('Geografia. Biografia. História' 
                                                                                                            if x < 1000 else ('Outros')))))))))))
```

#### Mudando o tipo de dado da coluna 'matricula_ou_siape' para string

```python
df.matricula_ou_siape = df.matricula_ou_siape.astype('string')
```
